# Cleanings

```python
# a function to clean the headers
def headers(x):
    x = x.drop(columns=['ID','Team & Contract','Loan Date End', 'Contract', 'Joined', 'Name'], axis=1)
    x.columns = x.columns.str.replace('[#,@,&,%,?,!,-,:]', '')
    x.columns = x.columns.str.replace(' ', '_')
    x.columns = x.columns.str.lower()
    x = x.drop(columns=['club',], axis=1)
    columns = ['ls','st','rs','lw','lf','cf','rf','rw','lam','cam','ram','lm','lcm','cm','rcm','rm','lwb','ldm', 'cdm','rdm','rwb','lb','lcb','cb','rcb','rb', 'gk']
    for col in columns:
        x[col] = x[col].str.split('+',n=1,expand = True)[0]
    return x
```

```python
# applying the function
fifa = headers(fifa)
```

```python
# replacing th special character from the end of height column
fifa['height'] = fifa['height'].str.rstrip('"')
```

```python
# applyng a conversion to the column, to turn feet&inches into cm and removing the other special character
convertions_heights = [30.48, 2.54]
fifa['height'] = fifa['height'].str.split("'").apply(pd.Series).astype(int).dot(convertions_heights)
```

```python
# removing lbs from the wheight
fifa['weight'] = fifa['weight'].str.rstrip('lbs')
```

```python
# converting lbs to kg
fifa["weight"] = pd.to_numeric(fifa['weight'])
# 0.454 number to convert lbs to kgs
fifa["weight"] = fifa["weight"].apply(lambda x: x*0.454) # ' lambda' to look up (Erin's tip)
```

```python
# Position column is just a group of other positions where the player can also play
# I believe it's not important to know in which positions a player can play to see it's overall hability
# It doesn't matter if your best position (bp) is 'striker' and you can play at 'left forward'...you are no
# better for playing it the left...you might be for playing in more than one position
# So...I'm changing the column of positions to a number of other positions where the player can play

# I'll fill nan's, on position column, to give them a value for the next operation (can't do ops on NaN's)
fifa['position'] = fifa['position'].fillna('no_other')
# Creating a new series with the different 'other positions' in a list, splitted by a space
other_positions = fifa['position'].str.split(' ')
```

```python
# creating a loop to count the number of other positions and to take the best position out the equation
size = []
for x in other_positions:
    z = len(x) - 1
    size.append(z)
```

```python
# appending this new list to the dataframe, resulting the a number of other positions were the player can also play
fifa['other_p'] = size
```

```python
# We don't need the column of positions anymore...
fifa = fifa.drop(columns = 'position', axis=1)
```

```python
def clean_numbers(x):
    value=[]
    for i in x:
        i = i.replace("€", "")
        if 'K' in i:
            i = i.replace("K","")
            i = float(i)
            i = float(i*1000)
            i = int(i)
            value.append(i)
        elif 'M' in i:
            i = i.replace("M","")
            i = float(i)
            i = float(i*1000000)
            i = int(i)
            value.append(i)
        else:
            value.append(i)
    return value
```

```python
# replacing characters in the columns wage, value and release clause.
fifa['wage'] = clean_numbers(fifa['wage'])
fifa['value'] = clean_numbers(fifa['value'])
fifa['release_clause'] = clean_numbers(fifa['release_clause'])
```

```python
fifa['wage'] = pd.to_numeric(fifa['wage'])
fifa['value'] = pd.to_numeric(fifa['value'])
fifa['release_clause'] = pd.to_numeric(fifa['release_clause'])
```

```python
#Removing the ★'s from the sm, w/f and ir columns
fifa['sm'] = fifa['sm'].str.rstrip('★')
fifa['w/f'] = fifa['w/f'].str.rstrip('★')
fifa['ir'] = fifa['ir'].str.rstrip('★')
```

```python
# these are the workrates for the attack and the defence...I will transform those into numbers
# since we only have 3 levels, I'll apply 1 for low, 2 for medium and 3 for high
fifa['a/w'] = fifa['a/w'].str.replace("High", '3')
fifa['a/w'] = fifa['a/w'].str.replace("Low", '1')
fifa['a/w'] = fifa['a/w'].str.replace("Medium", '2')
fifa['d/w'] = fifa['d/w'].str.replace("High", '3')
fifa['d/w'] = fifa['d/w'].str.replace("Low", '1')
fifa['d/w'] = fifa['d/w'].str.replace("Medium", '2')
```

```python
# now I can fill the NaN's with the median of those columns...
# filling NAN's with the median
fifa['a/w'] = fifa['a/w'].fillna(fifa['a/w'].median())
fifa['d/w'] = fifa['d/w'].fillna(fifa['d/w'].median())
```

```python
# Now I have to make a choice...
# I'm always into preserve the data, but after talking with Philip about the characteristics of this data,
# and after he told me that some of the players attributes are just a sum of others "Main Attributes"
# I thought I could fill the NaN's with the values that were missing, like, if
    #'attacking' = 300 (MAIN ATTRIBUTE)
        # crossing = 50
        #finishing = 75
        #heading_accuracy = 75
        #short_passing = 80
        #volleys = NaN
# So I would just fill the NaN with 80...to reach those 300 at the Main Attribute...
# But won't work because the Main Attribute is just a sum of the others attributes that are not NaN's...
# If I change the NaN's, I'll mess with the relationship of the Main Attributes...
# SO....let's drop NaN's...BTW, they are not significant if we take into account the number of entries that we have
# in this dataset...
```

```python
fifa = fifa.dropna()
```

```python
# when I was trying to transforme object columns into numeric dtypes, the following error appeard:
# Unable to parse string "3.3K" at position 213
# after checking this specific row (213)
# print(fifa.loc[[213]])
# I discovered that the problem was in the 'hits column' so I used the following code to see the unique values for this column
# fifa.hits.unique()
# and I discovered that some values have the K for the thousands and the dot...so I'll use the same code as before
```
